# Extracting more data 

##  Extract

```python
PAGE_TO_EXTRACT = 3
RAW_HTML_OUTPUT = '../data/raw_html'
CHORDS_DATASET_OUTPUT = f"../data/chords_{PAGE_TO_EXTRACT}.csv"
```

```python
import jl_io as io

genres = io.from_file('../data/genres')
decades = io.from_file('../data/decades')
```

```python
from jl_song_data import SongData

song_data = SongData()
```

```python
from jl_chord_scraping import ChordExtractor, LinkExtractor, ChordScraper

chord_extractor = ChordExtractor(RAW_HTML_OUTPUT)
link_extractor = LinkExtractor()
chord_scraper = ChordScraper(link_extractor,chord_extractor,genres,decades,song_data)
```

```python
chord_scraper.extract(first_page=PAGE_TO_EXTRACT, last_page=PAGE_TO_EXTRACT)
```

```python
song_data.df
```

```python
song_data.save(CHORDS_DATASET_OUTPUT)
```

## Cleaning 

```python
from jl_chord_cleaning import ChordDatasetCleaner
from jl_song_data import SongData
```

```python
RAW_HTML_OUTPUT = '../data/raw_html'
song_data = SongData(initial_data_path=CHORDS_DATASET_OUTPUT)
```

```python
song_data.df
```

```python
cleaner = ChordDatasetCleaner()
clean_dataset = cleaner.clean_dataset(song_data.df, RAW_HTML_OUTPUT)
```

```python
clean_dataset
```

```python
SongData(df=clean_dataset).save('../data/chords_clean_3.csv')
```

## Merge 

```python
data_old = SongData('../data/chords_clean_1_2.csv')
```

```python
data_old.df
```

```python
merged = cleaner.merge_datasets(data_old.df, clean_dataset)
```

```python
merged
```

```python
SongData(df=merged).save('../data/chords_clean_1_2_3.csv')
```
